```python
count = 0

output = "<table>"
json_output = {}
missingCount = 0
for anchor_article in outfits_payload:
#     count+=1
#     if count>100:
#         break
    
    articles = outfits_payload[anchor_article]
    print(articles)

    output += "<tr>"
    output += f"<td>{anchor_article}</td>"
    outfit = []
    for article in articles:
        
        
        file = f"./article_cache/article-{article}.json"

        if os.path.isfile(file):
            with open(file, "r") as read_file:
                response = json.load(read_file)        
        else:
            url = f"https://api.marksandspencer.com/phoenix-product/v1/articles/{article}?apikey={api_key}"
            response = requests.get(url).json()
            with open(file, "w") as write_file:
                json.dump(response, write_file)
        
        output += f"<td><a style=\"float:left;\" href=\"{response.get('carousel_image_url')}\">"
        output += f"<img src=\"{response.get('carousel_image_url')}\" style=\"width:100px;" 
        if anchor_article == article:
            output += 'border: 1px solid black;'

        output += f"\"><br/>{response.get('_id')}</a></td>"             
        
        # find macro category
        macroCategory = None
        for macro in macro_to_category:
            for category in macro_to_category[macro]:
                if category==response.get('category'):                    
                    macroCategory = macro
                    break;
        
        outfit.append({
            "articleId": article,
            "imageURL": response.get('carousel_image_url'),
            "priceRange": response.get('product_price_range'),
            "promotionalPriceRange": response.get('product_promotional_price_range'),
            "category": response.get('category'),
            "macroCategory": macroCategory
        })
        
        print(outfit)
    
    json_output[anchor_article] = outfit

    output += "</tr>"
output+="</table>"    
```
